```python
import pandas as pd
import numpy as np
import zipfile

with zipfile.ZipFile('titanic.zip', 'r') as zip_ref:
    zip_ref.extractall('.')


df = pd.read_csv("train.csv")


print(df.head())


df.info()


print(df.describe())


print("Shape:", df.shape)

print("\nMissing Values:")
print(df.isnull().sum())


numerical = df.select_dtypes(include=['int64', 'float64']).columns
print("\nNumerical Columns:")
print(numerical)

categorical = df.select_dtypes(include=['object']).columns
print("\nCategorical Columns:")
print(categorical)
```

The Titanic dataset contains 891 passengers and 12 features describing passenger information and survival status. Several columns contain missing values, especially Cabin, Age, and Embarked. The dataset includes both numerical features such as Age, Fare, and Passenger Class, as well as categorical features like Sex and Embarked. This initial exploration shows that data cleaning will be required before building any machine learning model. Overall, the dataset provides a solid foundation for learning data preprocessing and predictive modeling.
